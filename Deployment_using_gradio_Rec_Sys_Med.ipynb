```python

description = pd.read_csv('/content/description_data.csv')
precautions = pd.read_csv('/content/precautions_data.csv')
medications = pd.read_csv('/content/medications_data.csv')
diets = pd.read_csv('/content/diets_data.csv')
```

```python
# Prediction function
def get_predicted_value(patient_symptoms):
    input_vector = np.zeros(len(symptoms_dict))
    for symptom in patient_symptoms:
        if symptom in symptoms_dict:
            input_vector[symptoms_dict[symptom]] = 1

    # Create a DataFrame with the same column names used during training
    input_df = pd.DataFrame([input_vector], columns=symptoms_dict.keys())

    prediction = svc.predict(input_df)  # Use the DataFrame for prediction
    return disease_list.get(prediction[0], None)
```

```python
# Helper function to fetch details for a predicted disease
def helper(predicted_disease):
    if predicted_disease:
        desc = description.loc[description['Disease'] == predicted_disease, 'Description'].values
        desc = desc[0] if len(desc) > 0 else "No description available."

        prec = precautions.loc[precautions['Disease'] == predicted_disease, ['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
        prec = prec.values.flatten().tolist()

        med = medications.loc[medications['Disease'] == predicted_disease, 'Medication'].tolist()

        diet = diets.loc[diets['Disease'] == predicted_disease, 'Diet'].tolist()

        return desc, prec, med, diet
    else:
        return "Unknown disease", [], [], []
```

```python

# Function for Gradio
def predict_disease(symptoms):
    predicted_disease = get_predicted_value(symptoms)

    if predicted_disease is None:
        return "Sorry for the inconvenience, the disease you entered is not available in our database."

    desc, prec, med, diet = helper(predicted_disease)

    return (
        f"Predicted Disease: {predicted_disease}\n\n"
        f"Description: {desc}\n\n"
        f"Precautions: {','.join(prec) if prec else 'No precautions available.'}\n\n"
        f"Medications: {','.join(med) if med else 'No medications available.'}\n\n"
        f"Diet: {','.join(diet) if diet else 'No diet information available.'}"
    )
```

```python

# Create Gradio Interface
iface = gr.Interface(
    fn=predict_disease,
    inputs=gr.CheckboxGroup(choices=symptoms_list, label="Select your symptoms"),
    outputs="text",
    title="Disease Prediction System",
    description="Select your symptoms and the system will predict the disease and provide recommendations."
)
```
